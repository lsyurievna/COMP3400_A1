## Task 1 - Removing and Creating columns (15 marks)

For this task you need to show that you can handle simple pre-processing steps such as removing and creating new attributes in your dataframe. The tasks you must perform in this unit are listed below. 

1. Remove all columns to the right of WageEUR as we won't use them in this assignment.
2. Remove *Total Stats, BaseStats, Positions*, and *PhotoUrl*
3. Add An additional column labeled as GMValue (General Manager Value), that computes the following equation: $ 'Overall' * 0.9 + 'Growth'$, where overall and growth are columns present in your dataframe.
4. Add two other columns, one for first name (label it as FirstName), and one for last name (label it as LastName). 
5. Finally, sort your dataframe using GMValue in descending order

**Tip:** Print your dataframe in the end of this task and make sure it has 17 columns (this includes the three new ones created in this task). 


```python
#1 WageEUR is the 18th column, so iloc is used to select a slice of interest of the dataset
df = df.iloc[0:,0:18]

#2
to_drop = ['TotalStats','BaseStats', 'Positions', 'PhotoUrl']
df.drop(to_drop, inplace=True, axis = 1) #inplace True saves changes in df, not a copy

#3
df['GMValue'] = df['Overall']*0.9 + df['Growth']
```

```python

#4
#The names are split into fist and last names assuming that the first word in the full name
#is the first name, and the rest of the words are the person's last name
#This is done because some people have three or more words in their names, and the order
#of last name and first name is very cluture-specific.

#This method is not perfect as some players might have their middle name
#(along with any other non-last name words) included in their last name

#For example, the name David Ernesto Chicas Dueñas will be split into David as first name
#and Ernesto Chicas Dueñas as last name even though Ernesto is their middle name, and Chicas and Dueñas
#are two of their last names. 
df[['FirstName', 'LastName']] = df['FullName'].str.split(' ',1, expand = True)

#5
df.sort_values(by="GMValue", ascending = False)
```

## Task 3 - Evaluating groups of players (30 marks)

For task 3, we are interested in subgroups of players, rather than individuals.

1. What are the 5 nations with the highest average overall score and that at least 50 players are listed in the dataset?

    **Answer:** Brazil, The Chech Republic, Mexico, Portugal, Algeria


2. Which 5 clubs are paying the highest total wage to their players?

    **Answer:** England, Spain, France, Germany, Brazil


3. What is the average potential and average overall score for each group age younger than 19 (including 19 years old players)? Could you find any trend? Please discuss 

    **Answer:** As the player gets older, on average their overall score increases. Their potential, on the other hand, seems to decrease from 16 to 18 years old, with a slight increase from 18 to 19. 


```python
#1
agg_dict = {'Overall': ['mean','count']} 
new_df = df.groupby(['Nationality']).agg(agg_dict)
new_df = new_df[(new_df['Overall', 'count'] >= 50)] #selecting only the countries with at least 50 players
new_df.sort_values(by=('Overall','mean'), ascending = False).head(5) #sorting by average
```

```python
#2
agg_wages = {'WageEUR':['sum']}
wage_df = df.groupby('Nationality').agg(agg_wages)
wage_df.sort_values(by=('WageEUR','sum'),ascending=False).head(5)
```

```python
#3
age_df = df[(df['Age'] <= 19)]
agg_age = {'Potential':['mean'], 'Overall':['mean']}

young_df = age_df.groupby('Age').agg(agg_age)
young_df
```
